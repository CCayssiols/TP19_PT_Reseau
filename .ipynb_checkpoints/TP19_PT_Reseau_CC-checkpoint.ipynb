# TP : Formule des réseaux

```python
# commande permettant d'insérer les figures matplotlib dans les notebooks
%matplotlib widget

# import des librairies 
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

## Données LASER & Monte-Carlo

```python
lambda_LASER = 650e-9 # données constructeur JEULIN
u_lambda_LASER = 20e-9/np.sqrt(3) # données constructeur JEULIN
N = 100000 # nb de tirages aléatoires réalisés
```

## 530 traits / mm

```python
ordre_530 = np.array([-2,-1,0,1,2])
# distance_lat_530 = np.array([-8.8e-2, -3.5e-2, 0, 3.5e-2, 8.8e-2]) # valeurs Cyril
distance_lat_530 = np.array([-9.7e-2, -3.9e-2, 0, 4.0e-2, 9.8e-2]) # valeurs Christopher
u_distance_lat_530 = np.array([.2e-2/np.sqrt(3)]*len(distance_lat_530))
# distance_axiale_530 = 10e-2 # valeur Cyril
distance_axiale_530 = 10.9e-2 # valeur Christopher
u_distance_axiale_530 = .2e-2/np.sqrt(3)

sin_t_530 = []
u_sin_t_530 = []

for i in range(len(distance_lat_530)) :
    sin_store=[]
    for j in range(N):
        d = (distance_lat_530[i]+np.random.uniform(-1,1)*u_distance_lat_530[i])/(distance_axiale_530+np.random.uniform(-1,1)*u_distance_axiale_530)
        sin_store.append(np.sin(np.arctan(d)))
    sin_t_530.append(np.mean(sin_store))
    u_sin_t_530.append(np.std(sin_store, ddof=1))
```

```python
a_store = []
b_store = []
traits_mm_store = []

for i in range(N):
    z=np.polyfit(ordre_530, np.array(sin_t_530)+np.random.uniform(-1,1,len(sin_t_530))*np.array(u_sin_t_530), 1)
    a_store.append(z[0])
    b_store.append(z[1])
    traits_mm_store.append(1e-3/((lambda_LASER+u_lambda_LASER)/z[0]))

a_530 = np.mean(a_store)
b_530 = np.mean(b_store)
u_a_530 = np.std(a_store, ddof=1)
u_b_530 = np.std(b_store, ddof=1)
traits_mm_530 = np.mean(traits_mm_store)
u_traits_mm_530 = np.std(traits_mm_store, ddof=1)
```

```python
plt.figure()
plt.errorbar(ordre_530, sin_t_530, yerr=u_sin_t_530, fmt='o', label='sint_t moyen')
plt.plot(ordre_530, a_530*ordre_530+b_530, label='régression linéaire')
plt.xlabel('ordre')
plt.ylabel(r'$\sin t$')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("La pente mesurée est ", format(a_530, "#.3e"),"+/-",format(u_a_530, "#3e"))
print("L'ordonnée à l'origine mesurée est ", format(b_530, "#.3e"),"+/-",format(u_b_530, "#3e"))
print("Le nombre de traits par mm du réseau est de : ", format(traits_mm_530, "#.0f"), "+/-", format(u_traits_mm_530,"#.0f"),"traits / mm")
print("L'erreur relative est de ", format(100*np.abs((530-traits_mm_530)/traits_mm_530), "#.0f"), "%")
```
